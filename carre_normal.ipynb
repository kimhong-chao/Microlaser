```python
import meep as mp
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Initialized parameter 

b = 1        # base of the square (On peut prend l'unit√©) 10 pixel 
pml = 0.02     # the thickness of pml layer
pad = 0.01       # the padding between the cavity and the pml layer
n = 1.5       # refractive index the cavity (used resin)
resolution = 500
len_scale = 25   #in micrometer

sxy = b + 2*pml + 2*pad
cell_size = mp.Vector3(sxy, sxy, 0)

# Booudary conditions
boudary_layer = mp.PML(pml)

geometry = [mp.Block(mp.Vector3(b,b,mp.inf),
                     center=mp.Vector3(),
                     material=mp.Medium(index=n))]


# Injected sources
wl = 0.8                     # wavelenght in micrometer / in vaccum        
fcen = len_scale/wl         # pulse center frequency/ n refractive index
df = 15                 # pulse frequency width
src = mp.Source(mp.GaussianSource(fcen, fwidth=df), mp.Ez, mp.Vector3(0.25,0.1))

#Simulation
sim = mp.Simulation(cell_size= cell_size,
                    geometry= geometry,
                    sources=[src],
                    resolution=resolution,                    
                    boundary_layers=[boudary_layer])
```

```python
f = plt.figure(dpi=150)
sim.plot2D(ax = f.gca())
plt.show()
```

```python
# Chercher les fr√©quences de r√©sonnance autour de f = 0.15

h = mp.Harminv(mp.Ez, mp.Vector3(0.25,0.1), fcen, df)   # Est-ce qu'on s'int√©resse par les autres composants?

sim.run(mp.after_sources(h),
        until_after_sources=600)

# after_sources : time step in finte difference is half of the period of the source
```

```python
h2 = mp.Harminv(mp.Hx, mp.Vector3(0,0), fcen, df)   # Est-ce qu'on s'int√©resse par les autres composants?

sim.reset_meep()
sim.run(mp.after_sources(h2),
        until_after_sources=600)
```

```python
mode_Ez = np.array(h.modes)
mode_Hx = np.array(h2.modes)
res_modes = np.vstack((mode_Ez, mode_Hx))
```

```python
plt.plot(np.abs(res_modes[:,0]), np.abs(res_modes[:,1]), '+')
```
