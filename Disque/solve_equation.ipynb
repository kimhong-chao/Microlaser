```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import scipy.optimize as op
import scipy.special as sp
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
```

```python
def func(k, index, r, m) :
    return  np.abs(index*sp.jvp(m, index*k*r)/sp.jv(m, index*k*r)) - np.abs(sp.h1vp(m,k*r)/sp.hankel1(m,k*r))
```

```python
fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection='3d')

# Make data
u = np.linspace(0, 20, 500)
v = np.linspace(-1, 1 ,100)
X, Y = np.meshgrid(u, v)
# Plot the surface
surf = ax.plot_surface(X, Y, func(X+Y*1j,1.5,1,5), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-5, 100.01)
ax.zaxis.set_major_locator(LinearLocator(20))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('3D Plot of func $norm(left) - norm(right)$')
ax.set_xlabel('$Re(kR)$')
ax.set_ylabel('$Im(kR)$')
ax.set_zlabel('$norm(left) - norm(right)$')
plt.show()
plt.savefig('abs_func_en3D')
```
